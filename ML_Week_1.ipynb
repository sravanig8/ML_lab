```python
# Compute Central Tendency Measures: Mean, Median, Mode
# Measure of Dispersion: Variance, Standard Deviation with
# different sequences of data having real / integer /categorical
# /skewed. Also try with data having outliers
```

```python
import numpy as np
from statistics import mode
data = [10, 12, 14, 16, 18,18, 20, 22, 24, 26, 28]
print("Data:", data)
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Mode:", mode(data))
print("Variance:", np.var(data))
print("Standard Deviation:", np.std(data))
print("Range:", np.max(data) - np.min(data))
```

```python
from collections import Counter
data = ['Red', 'Blue', 'Red', 'Green', 'Red', 'Blue', 'Red']
mode_value = Counter(data).most_common(1)[0][0]
print("Data:", data)
print("Mode:", mode_value)
```
